## рҹ“– What is a Class? What is an Object?

A **class** is a **blueprint** вҖ” a template that defines what something is and what it can do.

An **object** is a **specific instance** created from that blueprint.

> **Analogy:**
> - **Class** = The architectural blueprint for a house
> - **Object** = An actual house built from that blueprint
> - You can build many houses from the same blueprint вҖ” each is a separate object

| Concept | Example |
|---------|---------|
| Class | `Student` (the concept of a student) |
| Object | `priya` (a specific student named Priya) |
| Class | `DataFrame` (Pandas' table concept) |
| Object | `df` (your specific table of sales data) |

## рҹ“– The `__init__` Method вҖ” Setting Up an Object

`__init__` is a special method that runs **automatically** when you create an object. It sets up the object's initial data.

```python
class ClassName:
    def __init__(self, param1, param2):
        self.param1 = param1
        self.param2 = param2
```

**`self`** refers to the specific object being created. Think of it as the object saying *"my name is..."*, *"my age is..."*.

````python
class Student:
    def __init__(self, name, age, course):
        self.name = name       # self.name belongs to THIS student
        self.age = age
        self.course = course

# Creating students вҖ” __init__ runs automatically
priya = Student("Priya", 24, "Data Science")
rahul = Student("Rahul", 28, "Machine Learning")

# Accessing their data
print(f"{priya.name} is {priya.age}, studying {priya.course}")
print(f"{rahul.name} is {rahul.age}, studying {rahul.course}")
````

````python
warning_box("Don't Forget self!",
    "Every method in a class must have <code>self</code> as its first parameter.<br>"
    "When you call <code>priya.some_method()</code>, Python automatically passes "
    "<code>priya</code> as <code>self</code>.<br><br>"
    "Forgetting <code>self</code> is the #1 beginner OOP mistake!")
````

## рҹ“– Methods вҖ” What an Object Can Do

A **method** is a function that belongs to a class. It can access the object's data using `self`.

````python
class Student:
    def __init__(self, name, scores):
        self.name = name
        self.scores = scores    # A list of exam scores
    
    def average(self):
        # Calculate the average score
        return sum(self.scores) / len(self.scores)
    
    def highest(self):
        # Return the highest score
        return max(self.scores)
    
    def is_passing(self, threshold=50):
        # Check if average is above the threshold
        return self.average() >= threshold

# Create a student
priya = Student("Priya", [85, 92, 78, 95, 88])

# Call methods вҖ” just like df.head() or model.fit()!
print(f"{priya.name}'s average: {priya.average():.1f}")
print(f"Highest score: {priya.highest()}")
print(f"Passing? {priya.is_passing()}")
print(f"Passing with 90+ average? {priya.is_passing(90)}")
````

## рҹ“– Making Objects Readable вҖ” `__repr__` and `__str__`

Right now, printing an object shows something ugly like `<__main__.Student object at 0x...>`. Let's fix that.

````python
class Student:
    def __init__(self, name, scores):
        self.name = name
        self.scores = scores
    
    def average(self):
        return sum(self.scores) / len(self.scores)
    
    def __repr__(self):
        # How the object shows up when printed or in a list
        return f"Student('{self.name}', avg={self.average():.1f})"
    
    def __str__(self):
        # A friendly string for print()
        return f"{self.name} вҖ” Average: {self.average():.1f}"

priya = Student("Priya", [85, 92, 78, 95])

print(priya)          # Uses __str__  вҶ’ friendly display
print(repr(priya))    # Uses __repr__ вҶ’ technical display
print([priya])        # Inside lists, __repr__ is used
````

## рҹҡҖ Practical Example: A Simple Dataset Class

Let's build something useful вҖ” a class that holds tabular data (like a mini Pandas DataFrame).

````python
class Dataset:
    # A simple tabular dataset вҖ” like a mini DataFrame.
    
    def __init__(self, data):
        # data: list of dictionaries (each dict = one row)
        self.data = data
    
    def num_rows(self):
        return len(self.data)
    
    def columns(self):
        if self.data:
            return list(self.data[0].keys())
        return []
    
    def get_column(self, col_name):
        # Extract all values from a single column
        return [row[col_name] for row in self.data]
    
    def filter(self, col_name, condition):
        # Filter rows where condition(value) is True
        filtered = [row for row in self.data if condition(row[col_name])]
        return Dataset(filtered)   # Return a NEW Dataset!
    
    def head(self, n=5):
        for row in self.data[:n]:
            print(row)
    
    def __repr__(self):
        return f"Dataset({self.num_rows()} rows, columns={self.columns()})"

# Let's use it!
sales = Dataset([
    {"product": "Laptop", "price": 75000, "quantity": 12},
    {"product": "Mouse", "price": 500, "quantity": 150},
    {"product": "Keyboard", "price": 1500, "quantity": 85},
    {"product": "Monitor", "price": 25000, "quantity": 30},
    {"product": "USB Cable", "price": 200, "quantity": 300},
])

print(sales)
print("Columns:", sales.columns())
print()
sales.head()
````

````python
# Using our Dataset methods вҖ” feels like Pandas!
prices = sales.get_column("price")
print("All prices:", prices)
print("Average price:", sum(prices) / len(prices))

# Filter expensive items (price > 1000)
expensive = sales.filter("price", lambda p: p > 1000)
print("\nExpensive items:")
print(expensive)
expensive.head()
````

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New balance: {self.balance}")
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds!")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
    
    def __repr__(self):
        return f"BankAccount('{self.owner}', balance={self.balance})"
```

```python
class ScoreTracker:
    def __init__(self, name):
        self.name = name
        self.scores = []
    
    def add_score(self, score):
        self.scores.append(score)
    
    def average(self):
        if not self.scores:
            return 0
        return sum(self.scores) / len(self.scores)
    
    def best(self):
        return max(self.scores) if self.scores else 0
    
    def summary(self):
        print(f"{self.name}: {len(self.scores)} scores, "
              f"avg={self.average():.1f}, best={self.best()}")
```
