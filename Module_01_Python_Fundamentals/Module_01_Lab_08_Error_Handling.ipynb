## 📖 Why Error Handling Matters

In the real world, things go wrong:
- A user types "abc" when you expect a number
- A file you're trying to open doesn't exist
- You divide by zero by accident
- A web server doesn't respond

Without error handling, your program **crashes** and stops. With error handling, your program can **recover gracefully** and keep running.

> **Analogy:** Error handling is like a **safety net** for a tightrope walker. The walker might not fall, but if they do, the net catches them instead of letting them hit the ground.

---
## 📖 Types of Errors

| Error Type | When it Happens | Example |
|-----------|----------------|---------|
| **SyntaxError** | Before code runs — you wrote invalid Python | Missing colon: `if x > 5` |
| **NameError** | Using a variable that doesn't exist | `print(xyz)` |
| **TypeError** | Wrong operation for the data type | `"hello" + 5` |
| **ValueError** | Right type but wrong value | `int("hello")` |
| **ZeroDivisionError** | Dividing by zero | `10 / 0` |
| **IndexError** | Accessing an index that doesn't exist | `[1,2,3][5]` |
| **KeyError** | Accessing a dictionary key that doesn't exist | `d["missing"]` |
| **FileNotFoundError** | Opening a file that doesn't exist | `open("fake.txt")` |

## 📖 try/except — Catching Errors

The `try/except` block lets you **attempt** something and **handle** it if it fails:

```python
try:
    # Code that MIGHT cause an error
    risky_operation()
except:
    # Code that runs IF an error occurs
    handle_the_error()
```

````python
# Without error handling — the program crashes
number = int(input("Enter a number: "))  # If user types "abc" → CRASH!

# With error handling — the program handles it gracefully
'''try:
    number = int(input("Enter a number: "))
    print(f"You entered: {number}")
except:
    print("❌ That's not a valid number! Please try again.")'''

print("Program continues running! ✅")
````

````python
# Catching SPECIFIC exception types
def safe_divide(a, b):
    """Divide a by b with error handling."""
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("❌ Cannot divide by zero!")
        return None
    except TypeError:
        print("❌ Both values must be numbers!")
        return None

print(safe_divide(10, 3))      # Works fine
print(safe_divide(10, 0))      # ZeroDivisionError caught
print(safe_divide(10, "abc"))  # TypeError caught
````

## 📖 The Full Pattern: try/except/else/finally

```python
try:
    # Code that might fail
except SomeError:
    # Runs ONLY if that error occurred
else:
    # Runs ONLY if NO error occurred
finally:
    # ALWAYS runs, whether error occurred or not
```

````python
# Full try/except/else/finally pattern
def get_number():
    try:
        text = input("Enter a number: ")
        number = float(text)
    except ValueError:
        print(f"❌ '{text}' is not a valid number!")
    else:
        print(f"✅ Successfully converted to {number}")
        return number
    finally:
        print("📌 (This always prints — cleanup goes here)")

result = get_number()
print("Result:", result)
````

````python
# Practical example: robust number input with retry
def get_valid_number(prompt, max_attempts=3):
    """Keep asking until user enters a valid number."""
    for attempt in range(1, max_attempts + 1):
        try:
            value = float(input(f"{prompt} (attempt {attempt}/{max_attempts}): "))
            return value  # Success — exit the function
        except ValueError:
            print("  ❌ Invalid input! Please enter a number.")
    
    print("  Too many failed attempts.")
    return None

# Try it:
result = get_valid_number("Enter your age")
if result is not None:
    print(f"Your age is: {result}")
else:
    print("Could not get a valid age.")
````

````python
exercise_header(2, "List Safe Access", "⭐⭐")

# ✏️ YOUR TURN: Write a function called 'safe_get' that:
# - Takes a list and an index
# - Returns the item at that index
# - If the index is out of range, returns a default value (None)
# - Hint: Catch IndexError

def safe_get(my_list, index, default=None):
    try:
        return my_list[index]
    except IndexError:
        return default

# Test it:
colors = ["red", "green", "blue"]
print(safe_get(colors, 0))    # "red"
print(safe_get(colors, 5))    # None
print(safe_get(colors, 10, "unknown"))  # "unknown" 
````
