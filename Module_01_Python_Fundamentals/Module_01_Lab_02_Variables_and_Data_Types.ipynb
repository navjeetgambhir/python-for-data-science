## 📖 What are Variables?

A **variable** is a named container that stores a value in the computer's memory.

> **Analogy:** Think of a variable as a **labelled box**. The label is the variable's name, and the contents of the box are the variable's value. You can:
> - Put something in the box (assign a value)
> - Look at what's in the box (use the variable)
> - Replace what's in the box with something new (reassign)

### Creating a Variable

To create a variable, you use the **assignment operator** `=` :

```python
variable_name = value
```

The `=` sign in Python does NOT mean "equals" like in maths. It means **"store this value in this variable"**. Think of it as an arrow pointing left: `variable_name ← value`.

````python
# Creating (declaring) a variable
message = "Hello, Python!"

# Using (accessing) the variable
print(message)
````

````python
# The variable stores the value — you can use it as many times as you want
greeting = "Good morning"

print(greeting)
print(greeting)
print(greeting, "everyone!")
````

````python
# You can change (reassign) a variable's value at any time
mood = "happy"
print("I am", mood)

mood = "excited"        # The old value "happy" is replaced
print(f"Now I am {mood}")
````

## 📖 Variable Naming Rules

Python has strict rules about what you can name your variables:

### ✅ Rules (You MUST follow these — or you'll get an error)

| Rule | Example ✅ | Bad Example ❌ |
|------|-----------|---------------|
| Must start with a letter or underscore | `name`, `_count` | `1name` (can't start with number) |
| Can contain letters, numbers, underscores | `score_1`, `player2` | `my-score` (no hyphens!) |
| Case-sensitive | `age` and `Age` are **different** variables | — |
| Cannot use Python keywords | `my_class` | `class` (reserved word) |

### ✅ Conventions (You SHOULD follow these — for readable code)

| Convention | Example | Why |
|-----------|---------|-----|
| Use `snake_case` | `student_name` | Standard Python style |
| Be descriptive | `total_sales` | Better than `ts` |
| Use lowercase | `age` | Capitals are for special cases |
| Avoid single letters | `score` not `s` | Exception: `i` in loops (later!) |

````python
# ✅ Good variable names
first_name = "Alice"
student_age = 25
total_score = 95.5
is_active = True

# ❌ Bad (but technically valid) names — avoid these!
x = "Alice"         # Not descriptive — what is x?
fn = "Alice"         # Abbreviation is unclear
FIRSTNAME = "Alice"  # All caps is reserved for constants

print("Good naming makes code readable:")
print("first_name:", first_name)
print("student_age:", student_age)
````

## 📖 Data Types — The Four Building Blocks

Every value in Python has a **type** that tells Python what kind of data it is.

Think of data types like different types of containers:
- A **text box** holds words and sentences → `str` (string)
- A **counter** holds whole numbers → `int` (integer)
- A **measuring cup** holds precise decimal numbers → `float`
- A **light switch** is either ON or OFF → `bool` (boolean)

| Type | Full Name | What It Holds | Examples |
|------|-----------|--------------|----------|
| `int` | Integer | Whole numbers (no decimals) | `42`, `-7`, `0`, `1000000` |
| `float` | Floating-point | Decimal numbers | `3.14`, `-0.5`, `100.0` |
| `str` | String | Text (letters, words, sentences) | `"hello"`, `'Python'`, `"123"` |
| `bool` | Boolean | Only `True` or `False` | `True`, `False` |

### 🔢 Integers (`int`) — Whole Numbers

Integers are whole numbers — no decimal point. They can be positive, negative, or zero.

````python
# Integers — whole numbers
age = 25
temperature = -3
population = 1400000000
zero = 0

print("Age:", age)
print("Temperature:", temperature)
print("Population:", population)
print("Zero:", zero)
print()
print("Type of age:", type(age))
````

````python
# You can do maths with integers
apples = 10
eaten = 3
remaining = apples - eaten

print("Started with:", apples)
print("Ate:", eaten)
print("Remaining:", remaining)
````

### 🔢 Floats (`float`) — Decimal Numbers

Floats are numbers with a decimal point. Use them when you need precision.

````python
# Floats — decimal numbers
pi = 3.14159
price = 29.99
percentage = 85.5
tiny = 0.001

print("Pi:", pi)
print("Price: $", price)
print("Score:", percentage, "%")
print("Type of pi:", type(pi))
````

````python
# Important: Adding .0 makes an integer into a float
whole = 5
decimal = 5.0

print(whole, "is type:", type(whole))    # int
print(decimal, "is type:", type(decimal))  # float

# They have the same VALUE but different TYPES
print("Are they equal?", whole == decimal)  # True!
````

### 📝 Strings (`str`) — Text

Strings hold text — any sequence of characters wrapped in quotation marks.

````python
# Strings — use single quotes OR double quotes (both work!)
name = "Alice"
greeting = 'Hello, World!'
address = "123 Main Street"
empty = ""  # An empty string (no characters)

print(name)
print(greeting)
print(address)
print("Empty string:", empty)
print("Type:", type(name))
````

````python
# When to use single vs double quotes?
# Use double quotes when your text contains an apostrophe:
sentence1 = "I'm learning Python"     # ✅ Works!
# sentence2 = 'I'm learning Python'   # ❌ Error! Python gets confused by the '

# Use single quotes when your text contains double quotes:
sentence3 = 'She said "hello" to me'  # ✅ Works!

print(sentence1)
print(sentence3)
````

````python
# IMPORTANT: Numbers in quotes are STRINGS, not numbers!
number = 42           # This is an integer
text_number = "42"    # This is a STRING — just text that looks like a number

print("number:", number, "| type:", type(number))
print("text_number:", text_number, "| type:", type(text_number))

# You can do maths with the integer...
print("number + 8 =", number + 8)

# ...but "adding" strings JOINS them (concatenation)!
print("text_number + '8' =", text_number + "8")  # "42" + "8" = "428" !
````

````python
warning_box("Strings vs Numbers",
    '<code>"42"</code> and <code>42</code> are COMPLETELY DIFFERENT to Python!<br>'
    '<code>"42"</code> is text — like the characters on a page.<br>'
    '<code>42</code> is a number — Python can do maths with it.<br>'
    'This is one of the most common sources of bugs for beginners.')
````

### ✅❌ Booleans (`bool`) — True or False

Booleans can only have one of two values: `True` or `False`.

They're used to represent yes/no, on/off, correct/incorrect situations. They become extremely important when we learn about `if` statements (making decisions) in the next notebook.

> **Analogy:** A boolean is like a light switch — it's either ON (`True`) or OFF (`False`). There's no in-between.

````python
# Booleans — only True or False (capital T and F!)
is_student = True
has_graduated = False

print("Is a student?", is_student)
print("Has graduated?", has_graduated)
print("Type:", type(is_student))
````

````python
# Booleans come from comparisons
age = 20

print("age > 18?", age > 18)      # True (20 is greater than 18)
print("age == 25?", age == 25)     # False (20 is not equal to 25)
print("age < 30?", age < 30)      # True
print("age >= 20?", age >= 20)    # True (20 is greater than or equal to 20)
````

````python
warning_box("True/False are Case-Sensitive!",
    "You must write <code>True</code> and <code>False</code> with a capital first letter.<br>"
    "<code>true</code> and <code>false</code> (lowercase) will cause an error!")
````

## 📖 Checking Types with `type()`

The `type()` function tells you what data type a value or variable is. This is incredibly useful when debugging!

````python
# Use type() to check what kind of data you have
print(type(42))          # <class 'int'>
print(type(3.14))        # <class 'float'>
print(type("hello"))     # <class 'str'>
print(type(True))        # <class 'bool'>
````

````python
# You can use type() on variables too
mystery = "Python"
print("The type of mystery is:", type(mystery))

# Let's reassign and check again
mystery = 100
print("Now the type of mystery is:", type(mystery))
````

## 📖 Type Casting — Converting Between Types

Sometimes you need to convert a value from one type to another. This is called **type casting** or **type conversion**.

Common conversions:

| Function | What it does | Example |
|----------|-------------|---------|
| `int()` | Converts to integer | `int("42")` → `42` |
| `float()` | Converts to float | `float("3.14")` → `3.14` |
| `str()` | Converts to string | `str(42)` → `"42"` |
| `bool()` | Converts to boolean | `bool(1)` → `True` |

````python
# Converting strings to numbers
age_text = "25"               # This is a string
age_number = int(age_text)    # Convert to integer

print("age_text:", age_text, "| type:", type(age_text))
print("age_number:", age_number, "| type:", type(age_number))

# Now we can do maths!
print("In 5 years:", age_number + 5)
````

````python
# This is VERY common with input() — it always returns a string!
# Even if the user types a number, input() gives you a STRING.

# Without conversion:
# user_age = input("Enter your age: ")
# print(user_age + 10)  # ❌ ERROR! Can't add string + number

# With conversion:
# user_age = int(input("Enter your age: "))
# print(user_age + 10)  # ✅ Works!

# Let's demonstrate without input():
user_typed = "30"                   # Simulating what input() returns
user_age = int(user_typed)          # Convert string to integer
print("Age in 10 years:", user_age + 10)
````

````python
# Converting numbers to strings (useful for building messages)
score = 95
message = "Your score is: " + str(score) + " out of 100"
print(message)
````

````python
# Converting to float
whole_number = 42
as_float = float(whole_number)
print(whole_number, "→", as_float)

text_decimal = "3.14"
as_float2 = float(text_decimal)
print(text_decimal, "→", as_float2)
````

````python
# Boolean conversions — what counts as True vs False?
# These are "falsy" (become False):
print("bool(0):", bool(0))
print('bool(""):', bool(""))       # Empty string
print("bool(0.0):", bool(0.0))
print("bool(None):", bool(None))

print()

# Everything else is "truthy" (becomes True):
print("bool(1):", bool(1))
print("bool(42):", bool(42))
print("bool(-1):", bool(-1))
print('bool("hello"):', bool("hello"))
print("bool(0.1):", bool(0.1))
````

````python
info_box("The input() Trap",
    "The <code>input()</code> function ALWAYS returns a <b>string</b>, even if the user types a number.<br>"
    "If you need a number, you must convert it: <code>age = int(input('Enter age: '))</code><br>"
    "This is one of the most common beginner mistakes in Python!")
````

````python
# ❌ Common error: trying to convert text that isn't a valid number
# int("hello")       # ValueError: invalid literal for int()
# int("3.14")        # ValueError: can't convert decimal string directly to int

# ✅ To convert "3.14" to int, go through float first:
result = int(float("3.14"))
print("'3.14' as int:", result)  # 3 (decimal part is dropped)
````
