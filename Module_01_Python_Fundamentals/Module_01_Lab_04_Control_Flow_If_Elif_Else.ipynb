## 📖 Why Do We Need Control Flow?

So far, Python has executed every line of our code in order, from top to bottom. But real programs need to **make decisions**:

- *If* the user is over 18, show adult content
- *If* the temperature is below 0, display a frost warning
- *If* the password is correct, log in; *otherwise*, show an error

> **Analogy:** Control flow is like a **road junction**. At a junction, you choose which road to take based on where you want to go. In Python, you choose which code to run based on a condition.

The `if` statement is how Python makes decisions.

---
## 📖 The `if` Statement — Basic Decisions

The simplest form of decision-making:

```python
if condition:
    # This code runs ONLY if condition is True
    do_something()
```

**Two critical rules:**
1. The `if` line must end with a **colon** `:`
2. The code inside must be **indented** (pushed to the right with spaces/tab)

````python
# Simple if statement
age = 20

if age >= 18:
    print("You are an adult!")

print("This line runs regardless of age")
````

````python
# The condition is checked — if True, the indented code runs
temperature = 35

if temperature > 30:
    print("It's a hot day!")
    print("Stay hydrated!")   # Both lines are indented → both are inside the if

print("Temperature check complete.")  # Not indented → runs always
````

````python
# If the condition is False, the indented code is SKIPPED
temperature = 20

if temperature > 30:
    print("It's a hot day!")    # This does NOT run (20 is not > 30)
    print("Stay hydrated!")     # This does NOT run either

print("Temperature check complete.")  # This DOES run
````

````python
warning_box("Indentation is NOT Optional!",
    "In many programming languages, indentation is just for readability. "
    "In Python, indentation is <b>part of the syntax</b> — it tells Python which lines "
    "belong inside the <code>if</code> block.<br><br>"
    "Standard: Use <b>4 spaces</b> for each level of indentation.<br>"
    "Jupyter notebooks handle this automatically when you press Tab.")
````

## 📖 The `if/else` Statement — Two Paths

What if you want to do one thing when the condition is True, and a **different thing** when it's False?

```python
if condition:
    # Runs when condition is True
else:
    # Runs when condition is False
```

> **Analogy:** It's like a fork in the road — you must go left OR right, never both.

````python
# if/else — choosing between two options
age = 15

if age >= 18:
    print("You can vote! 🗳️")
else:
    print("Sorry, you're too young to vote.")
    years_left = 18 - age
    print(f"You can vote in {years_left} years.")
````

````python
# Another example
number = 7

if number % 2 == 0:
    print(number, "is EVEN")
else:
    print(number, "is ODD")
````

## 📖 The `if/elif/else` Statement — Multiple Paths

When you have more than two options, use `elif` (short for "else if"):

```python
if condition1:
    # Runs if condition1 is True
elif condition2:
    # Runs if condition1 is False AND condition2 is True
elif condition3:
    # Runs if condition1 and condition2 are False AND condition3 is True
else:
    # Runs if ALL conditions above are False
```

> **Analogy:** It's like a multiple-choice question — Python checks each option from top to bottom and picks the **first one** that's True.

````python
# Grading system with if/elif/else
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} → Grade: {grade}")
````

````python
info_box("Order Matters!",
    "Python checks conditions <b>from top to bottom</b> and stops at the first True one.<br>"
    "If you wrote <code>score >= 60</code> before <code>score >= 90</code>, a score of 95 would "
    "match <code>>= 60</code> first and get a D!<br><br>"
    "Always put the most specific/restrictive conditions FIRST.")
````

## 📖 Nested `if` Statements — Decisions Within Decisions

You can put an `if` statement inside another `if` statement. This is called **nesting**.

```python
if condition1:
    if condition2:
        # Runs if BOTH condition1 AND condition2 are True
```

> **Analogy:** It's like going through security at an airport. First, they check your ticket. If you have a ticket, THEN they check your ID. Both checks must pass.

````python
# Nested if — checking multiple levels
has_ticket = True
age = 15

if has_ticket:
    print("✅ Ticket verified!")
    if age >= 18:
        print("🎬 Enjoy the R-rated movie!")
    else:
        print("🚫 Sorry, you must be 18+ for this movie.")
else:
    print("❌ You need a ticket first!")
````

````python
success_box(
    "You can often replace nested <code>if</code> statements with <code>and</code>:<br>"
    "<code>if has_ticket and age >= 18:</code><br>"
    "This is usually cleaner. Use nesting only when the inner logic is complex.")
````

## 🎯 Key Takeaways

1. **`if`** runs code only when a condition is True
2. **`if/else`** chooses between two paths — one for True, one for False
3. **`if/elif/else`** chooses between multiple paths — first True condition wins
4. **Indentation** (4 spaces) is how Python knows which code belongs inside a block
5. Conditions are checked **top to bottom** — order matters!
