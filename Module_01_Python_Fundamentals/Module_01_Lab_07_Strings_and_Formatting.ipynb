## 📖 Indexing — Accessing Individual Characters

Each character in a string has a position number called an **index**. Python counts from **0**.

```
 P  y  t  h  o  n
 0  1  2  3  4  5     ← positive index
-6 -5 -4 -3 -2 -1    ← negative index (counting from the end)
```

````python
word = "Python"

# Positive indexing (from the start)
print("First character:", word[0])      # P
print("Second character:", word[1])     # y
print("Last character:", word[5])       # n

# Negative indexing (from the end)
print("Last character:", word[-1])      # n
print("Second to last:", word[-2])      # o
````

## 📖 Slicing — Extracting Parts of a String

Slicing extracts a portion of a string: `string[start:stop:step]`

- `start`: Where to begin (included)
- `stop`: Where to end (NOT included)
- `step`: How many characters to skip

````python
text = "Data Science"

print("First 4 characters:", text[0:4])   # "Data"
print("Characters 5-11:", text[5:12])      # "Science"
print("From index 5 to end:", text[5:])    # "Science"
print("First 4:", text[:4])               # "Data" (start defaults to 0)
print("Every 2nd character:", text[::2])   # "Dt cec"
print("Reversed:", text[::-1])            # "ecneicS ataD" 
````

## 📖 String Methods — Built-in Superpowers

Strings come with many useful methods (functions that belong to the string). Call them with a **dot**: `string.method()`

| Method | What it does | Example |
|--------|-------------|---------|
| `.upper()` | All uppercase | `"hello".upper()` → `"HELLO"` |
| `.lower()` | All lowercase | `"HELLO".lower()` → `"hello"` |
| `.title()` | Capitalise each word | `"hello world".title()` → `"Hello World"` |
| `.strip()` | Remove leading/trailing spaces | `" hello ".strip()` → `"hello"` |
| `.replace(old, new)` | Replace text | `"cat".replace("c", "b")` → `"bat"` |
| `.split(sep)` | Split into a list | `"a,b,c".split(",")` → `["a","b","c"]` |
| `.join(list)` | Join a list into string | `",".join(["a","b"])` → `"a,b"` |
| `.find(sub)` | Find position of substring | `"hello".find("ll")` → `2` |
| `.count(sub)` | Count occurrences | `"hello".count("l")` → `2` |
| `.startswith(s)` | Check if starts with | `"hello".startswith("he")` → `True` |
| `.endswith(s)` | Check if ends with | `"hello".endswith("lo")` → `True` |

````python
# Case conversion
text = "python for Data Science"
print("Original:", text)
print("Upper:", text.upper())
print("Lower:", text.lower())
print("Title:", text.title())
print("Capitalize:", text.capitalize())
````

````python
# Stripping whitespace (very common in data cleaning!)
messy = "   Hello, World!   "
print(f"Before strip: '{messy}'")
print(f"After strip:  '{messy.strip()}'")
print(f"Left strip:   '{messy.lstrip()}'")
print(f"Right strip:  '{messy.rstrip()}'")
````

````python
# Replace
sentence = "I love cats. Cats are the best. CATS!"
print("Original:", sentence)
print("Replace 'cats' with 'dogs':", sentence.replace("cats", "dogs"))
# Note: replace is case-sensitive! "Cats" and "CATS" were NOT replaced.
# To replace all: convert to lower first
print("Replace all:", sentence.lower().replace("cats", "dogs"))
````

````python
# Split and Join — VERY useful in Data Science
csv_row = "Alice,25,Mumbai,Data Science"
fields = csv_row.split(",")
print("Split result:", fields)
print("Name:", fields[0])
print("Age:", fields[1])
print("City:", fields[2])

# Join — the reverse of split
words = ["Python", "is", "awesome"]
sentence = " ".join(words)
print("Joined:", sentence)
````

````python
# Checking string properties
print("'hello123'.isalnum():", "hello123".isalnum())    # True (letters + digits)
print("'hello'.isalpha():", "hello".isalpha())           # True (only letters)
print("'12345'.isdigit():", "12345".isdigit())           # True (only digits)
print("'hello'.islower():", "hello".islower())           # True
print("'HELLO'.isupper():", "HELLO".isupper())           # True
print("'  '.isspace():", "  ".isspace())                 # True
````

## 📖 f-Strings — Modern String Formatting

f-strings (formatted string literals) are the best way to build strings with variables. Put `f` before the opening quote and use `{variable}` inside:

````python
# f-strings — clean and readable
name = "Alice"
age = 25
city = "Mumbai"

# Old way (concatenation) — messy!
print("My name is " + name + ", I am " + str(age) + " years old, from " + city + ".")

# New way (f-string) — clean!
print(f"My name is {name}, I am {age} years old, from {city}.")
````

````python
# f-strings can contain expressions
price = 49.99
quantity = 3
print(f"Total: ₹{price * quantity}")
print(f"5 squared is {5**2}")
print(f"Is 10 > 5? {10 > 5}")
````

````python
# Number formatting with f-strings
pi = 3.14159265

print(f"Pi: {pi}")               # Full precision
print(f"Pi (2 decimals): {pi:.2f}")  # 3.14
print(f"Pi (4 decimals): {pi:.4f}")  # 3.1416

big_number = 1234567890
print(f"Formatted: {big_number:,}")  # 1,234,567,890

percentage = 0.856
print(f"Percentage: {percentage:.1%}")  # 85.6%
````
