## 📖 Arithmetic Operators — Doing Maths

Python can work as a powerful calculator. Here are the arithmetic operators:

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `10 + 3` | `13` |
| `-` | Subtraction | `10 - 3` | `7` |
| `*` | Multiplication | `10 * 3` | `30` |
| `/` | Division | `10 / 3` | `3.333...` |
| `//` | Floor Division (integer division) | `10 // 3` | `3` |
| `%` | Modulo (remainder) | `10 % 3` | `1` |
| `**` | Exponentiation (power) | `10 ** 3` | `1000` |

````python
# Basic arithmetic — Python as a calculator
print("Addition:       10 + 3 =", 10 + 3)
print("Subtraction:    10 - 3 =", 10 - 3)
print("Multiplication: 10 * 3 =", 10 * 3)
print("Division:       10 / 3 =", 10 / 3)
````

````python
# Floor division — divides and drops the decimal part
print("Floor division: 10 // 3 =", 10 // 3)  # 3 (not 3.33)
print("Floor division: 17 // 5 =", 17 // 5)  # 3

# Modulo — gives the REMAINDER after division
print("Modulo:         10 % 3 =", 10 % 3)    # 1 (because 10 = 3*3 + 1)
print("Modulo:         17 % 5 =", 17 % 5)    # 2 (because 17 = 5*3 + 2)

# Exponentiation — raises to a power
print("Power:          2 ** 3 =", 2 ** 3)     # 8 (2 × 2 × 2)
print("Power:          5 ** 2 =", 5 ** 2)     # 25 (5 × 5)
````

````python
info_box("Floor Division vs Regular Division",
    "<code>10 / 3 = 3.3333</code> → Regular division always gives a float (decimal result)<br>"
    "<code>10 // 3 = 3</code> → Floor division drops the decimal part (rounds DOWN)<br><br>"
    "<b>Data Science use:</b> Modulo (%) is great for checking if a number is even/odd:<br>"
    "<code>number % 2 == 0</code> means the number is even!")
````

````python
# Practical example: Is a number even or odd?
number = 42
remainder = number % 2
print(number, "divided by 2 has remainder:", remainder)

if remainder == 0:
    print(number, "is EVEN")
else:
    print(number, "is ODD")
````

````python
# Division ALWAYS returns a float, even if the result is whole
print("10 / 2 =", 10 / 2)          # 5.0 (float, not int!)
print("Type:", type(10 / 2))        # <class 'float'>

# Floor division returns an int (when both operands are ints)
print("10 // 2 =", 10 // 2)        # 5 (int)
print("Type:", type(10 // 2))       # <class 'int'>
````

## 📖 Comparison Operators — Comparing Values

Comparison operators compare two values and return a **boolean** (`True` or `False`).

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `5 > 3` | `True` |
| `<` | Less than | `5 < 3` | `False` |
| `>=` | Greater than or equal to | `5 >= 5` | `True` |
| `<=` | Less than or equal to | `5 <= 3` | `False` |

````python
# Comparing numbers
print("5 == 5:", 5 == 5)    # True — are they equal?
print("5 == 3:", 5 == 3)    # False
print("5 != 3:", 5 != 3)    # True — are they NOT equal?
print("5 > 3:", 5 > 3)      # True
print("5 < 3:", 5 < 3)      # False
print("5 >= 5:", 5 >= 5)    # True — greater than OR equal to
print("5 <= 3:", 5 <= 3)    # False
````

````python
warning_box("= vs ==",
    "<code>=</code> is <b>assignment</b>: <code>x = 5</code> means 'store 5 in x'<br>"
    "<code>==</code> is <b>comparison</b>: <code>x == 5</code> means 'is x equal to 5?'<br><br>"
    "This is one of the most common mistakes! If you use <code>=</code> when you mean "
    "<code>==</code>, Python will assign a value instead of comparing.")
````

````python
# Comparing strings
print('"apple" == "apple":', "apple" == "apple")    # True
print('"apple" == "Apple":', "apple" == "Apple")    # False! (case-sensitive)
print('"apple" < "banana":', "apple" < "banana")    # True (alphabetical order)
````

````python
# Using comparisons with variables
age = 20
can_vote = age >= 18
print("Age:", age)
print("Can vote?", can_vote)  # True

temperature = 38.5
has_fever = temperature > 37.5
print("Temperature:", temperature)
print("Has fever?", has_fever)  # True
````

## 📖 Logical Operators — Combining Conditions

Logical operators let you combine multiple conditions:

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `and` | Both must be True | `True and True` | `True` |
| `or` | At least one must be True | `True or False` | `True` |
| `not` | Reverses the value | `not True` | `False` |

> **Analogy:**
> - `and` is like a **strict parent** — "You can go out IF you finish homework AND clean your room" (both needed)
> - `or` is like a **relaxed parent** — "You can have ice cream OR cake" (at least one)
> - `not` is like saying **"the opposite"** — "not raining" means it's dry

````python
# AND — both conditions must be True
age = 25
has_license = True

can_drive = age >= 18 and has_license
print("Can drive?", can_drive)  # True (both conditions met)

age = 15
can_drive = age >= 18 and has_license
print("Can drive (age 15)?", can_drive)  # False (age condition fails)
````

````python
# OR — at least one condition must be True
is_weekend = True
is_holiday = False

is_day_off = is_weekend or is_holiday
print("Day off?", is_day_off)  # True (weekend is True)

is_weekend = False
is_day_off = is_weekend or is_holiday
print("Day off (no weekend, no holiday)?", is_day_off)  # False
````

````python
# NOT — reverses True/False
is_raining = False
go_outside = not is_raining
print("Is it raining?", is_raining)
print("Should we go outside?", go_outside)  # True (not raining → go out!)
````

````python
# Combining multiple operators
age = 25
income = 50000
has_good_credit = True

# Eligible for loan: age >= 21 AND (income >= 40000 OR good credit)
eligible = age >= 21 and (income >= 40000 or has_good_credit)
print("Loan eligible?", eligible)  # True
````

## 📖 Assignment Operators — Shortcuts

Python provides shortcut operators for common operations:

| Shortcut | Equivalent to | Meaning |
|----------|--------------|---------|
| `x += 5` | `x = x + 5` | Add 5 to x |
| `x -= 3` | `x = x - 3` | Subtract 3 from x |
| `x *= 2` | `x = x * 2` | Multiply x by 2 |
| `x /= 4` | `x = x / 4` | Divide x by 4 |
| `x //= 3` | `x = x // 3` | Floor divide x by 3 |
| `x %= 2` | `x = x % 2` | x modulo 2 |
| `x **= 3` | `x = x ** 3` | x to the power 3 |

````python
# Assignment operators in action
score = 100
print("Starting score:", score)

score += 10    # Same as: score = score + 10
print("After +10:", score)

score -= 25    # Same as: score = score - 25
print("After -25:", score)

score *= 2     # Same as: score = score * 2
print("After ×2:", score)

score //= 3    # Same as: score = score // 3
print("After ÷3 (floor):", score)
````

## 📖 Operator Precedence — Order of Operations

Just like in maths (BODMAS/PEMDAS), Python follows a specific order:

| Priority | Operator | Description |
|----------|----------|-------------|
| 1 (highest) | `()` | Parentheses (always first) |
| 2 | `**` | Exponentiation |
| 3 | `*`, `/`, `//`, `%` | Multiplication, division |
| 4 | `+`, `-` | Addition, subtraction |
| 5 | `==`, `!=`, `<`, `>`, `<=`, `>=` | Comparisons |
| 6 | `not` | Logical NOT |
| 7 | `and` | Logical AND |
| 8 (lowest) | `or` | Logical OR |

> **Rule of thumb:** When in doubt, use **parentheses** `()` to make your intention clear!

````python
# Order matters!
print("Without parentheses: 2 + 3 * 4 =", 2 + 3 * 4)    # 14 (not 20!)
print("With parentheses:    (2 + 3) * 4 =", (2 + 3) * 4)  # 20

print()
print("Without: 10 - 2 ** 2 =", 10 - 2 ** 2)    # 6 (** first: 2²=4, then 10-4=6)
print("With:    (10 - 2) ** 2 =", (10 - 2) ** 2)  # 64
````
