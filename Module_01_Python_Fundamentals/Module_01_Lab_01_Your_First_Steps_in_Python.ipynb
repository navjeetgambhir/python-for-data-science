## 📖 Using `print()` — Displaying Output

`print()` is one of the most useful functions in Python. You will use it constantly to:
- See the results of your calculations
- Check what your program is doing (called **debugging**)
- Display information to the user

Let's explore different things you can print:

```python
# You can print any text (string) — just put it in quotes
print("Welcome to Python for Data Science!")
```

```python
# You can print numbers — no quotes needed for numbers
print(42)
```

```python
# You can print the result of calculations
print(10 + 5)
```

```python
# You can print multiple things by separating them with commas
print("The answer is:", 42)
```

```python
# You can print multiple lines by using multiple print statements
print("Line 1: Hello!")
print("Line 2: How are you?")
print("Line 3: Let's learn Python!")
```

```python
# An empty print() just creates a blank line
print("Before the gap")
print()
print("After the gap")
```

## 📖 Comments — Notes for Humans

You've already seen lines starting with `#` in the code above. These are called **comments**.

**What are comments?**
- Lines that Python **completely ignores** — they are NOT instructions
- They are notes written **for humans** (you, your teammates, your future self)
- They help explain **what** the code does and **why**

**Why use comments?**
- When you come back to your code weeks later, comments remind you what you were thinking
- When someone else reads your code, comments help them understand it
- In Data Science, comments explain your analysis decisions

> **Analogy:** Comments are like sticky notes on a recipe. The oven doesn't read the sticky notes, but they remind *you* why you added extra garlic last time!

```python
# This entire line is a comment — Python ignores it completely
# I can write anything here and Python won't care

print("This line runs!")  # Comments can also go at the END of a line

# The print below is "commented out" — it won't run because of the #
# print("This line does NOT run because it starts with #")
```

```python
# EXPERIMENT: Uncomment the line below by removing the # at the start.
# Then run this cell to see the difference.

print("I was hidden! Now I'm visible!")
```

### Multi-line Comments

If you need to write a longer note spanning multiple lines, you have two options:

```python
# Option 1: Use # on each line (most common)
# This is a comment
# that spans multiple
# lines using hash symbols

# Option 2: Use triple quotes (less common for comments, but works)
"""
This is also a multi-line comment.
Python technically sees this as a string,
but since we don't assign it to anything,
it gets ignored.
"""

print("Comments explained!")
```

## 📖 Getting Input from the User — `input()`

So far, we've been *giving* information to the user with `print()`. But what if we want to *get* information from the user?

That's what `input()` does — it:
1. Displays a message (called a **prompt**) to the user
2. Waits for the user to type something and press Enter
3. Returns whatever the user typed

> **Analogy:** `print()` is like a speaker — it broadcasts information out. `input()` is like a microphone — it receives information in.

```python
# When you run this cell, it will ask you to type your name.
# Type your name in the box that appears, then press Enter.

name = input("What is your name? ")
print("Hello,", name, "— welcome to Python!")
```

### What just happened?

Let's break that code down step by step:

| Code | What it does |
|------|-------------|
| `input("What is your name? ")` | Shows the message and waits for you to type |
| `name = ` | **Stores** whatever you typed into a container called `name` |
| `print("Hello,", name, ...)` | Displays a greeting using the name you typed |

The `name` part is a **variable** — we'll learn all about variables in the next notebook. For now, just think of it as a labelled box that holds a value.

```python
# Another example — asking for a favourite colour
colour = input("What is your favourite colour? ")
print("Great choice!", colour, "is a lovely colour!")
```

## ⚠️ Common Beginner Mistakes

Mistakes are **normal** and **expected** in programming! Every professional programmer makes errors every single day. The key is learning to **read error messages** — Python tries to tell you what went wrong.

Let's look at some common mistakes beginners make:

### Mistake 1: Forgetting the quotation marks

```python
# ❌ This will cause an error — run it and read the error message!
# (Don't worry, errors won't break your computer!)

# print(Hello World)

# ✅ The fix: Add quotation marks around text
print("Hello World")
```

### Mistake 2: Mismatched quotation marks

```python
# ❌ This will cause an error — starting with " but ending with '
# print("Hello World')

# ✅ The fix: Use matching quotes (both " or both ')
print("Hello World")
print('Hello World')  # Single quotes work too!
```

### Mistake 3: Forgetting the parentheses

```python
# ❌ This will cause an error in Python 3
# print "Hello"

# ✅ The fix: Always use parentheses with print()
print("Hello")
```

### Mistake 4: Capital letters matter!

```python
# ❌ Python is case-sensitive — Print is not the same as print
# Print("Hello")   # This would cause an error

# ✅ The fix: Use lowercase 'print'
print("Hello")
```

```python
warning_box("Remember!", 
    "Python is <b>case-sensitive</b>. That means <code>print</code>, <code>Print</code>, "
    "and <code>PRINT</code> are all different things to Python. Always use lowercase <code>print</code>.")
```
