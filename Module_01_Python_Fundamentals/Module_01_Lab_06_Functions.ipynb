## 📖 What are Functions?

A **function** is a reusable block of code that performs a specific task.

> **Analogy:** A function is like a **recipe**. You give it ingredients (inputs), it follows steps, and it gives you back a dish (output). Once you write the recipe, you can use it any time without writing all the steps again.

You've already used built-in functions: `print()`, `input()`, `type()`, `int()`, `len()`. Now you'll learn to **create your own!**

### Why Use Functions?

| Benefit | Explanation |
|---------|-------------|
| **Avoid repetition** | Write once, use many times |
| **Organised code** | Break big problems into small pieces |
| **Easier testing** | Test each piece independently |
| **Readability** | `calculate_tax(income)` is clearer than 10 lines of maths |

## 📖 Defining a Function

```python
def function_name(parameters):
    """Docstring — describes what the function does."""
    # Function body — the code that runs
    return result  # Optional — sends a value back
```

| Part | What it is |
|------|-----------|
| `def` | Keyword that says "I'm defining a function" |
| `function_name` | The name you give your function (use snake_case) |
| `parameters` | Input variables (in parentheses) — optional |
| `"""Docstring"""` | A description of what the function does |
| `return` | Sends a result back — optional |

````python
# Defining a simple function (no parameters, no return)
def say_hello():
    """Print a friendly greeting."""
    print("Hello! Welcome to Python! 👋")

# IMPORTANT: Defining a function does NOT run it.
# You must CALL it by adding parentheses:
say_hello()      # This runs the function
say_hello()      # You can call it as many times as you want
````

````python
# Function with a parameter
def greet(name):
    """Greet a person by name."""
    print(f"Hello, {name}! Welcome to Python! 👋")

greet("Alice")     # "Alice" is the ARGUMENT — the actual value we pass in
greet("Bob")
greet("Charlie")
````

````python
info_box("Parameter vs Argument",
    "<b>Parameter:</b> The variable name in the function definition: <code>def greet(name)</code><br>"
    "<b>Argument:</b> The actual value you pass when calling: <code>greet('Alice')</code><br><br>"
    "Think of it this way: the parameter is the parking space, the argument is the car.")
````

````python
# Function with multiple parameters
def introduce(name, age, city):
    """Introduce a person with their details."""
    print(f"Hi! I'm {name}, I'm {age} years old, and I live in {city}.")

introduce("Priya", 25, "Mumbai")
introduce("Rahul", 30, "Delhi")
````

## 📖 Return Values — Getting Results Back

So far our functions just printed things. But often you want a function to **calculate something and give it back** to you. That's what `return` does.

````python
# Function that RETURNS a value
def add(a, b):
    """Add two numbers and return the result."""
    result = a + b
    return result

# Capture the returned value in a variable
total = add(10, 5)
print("10 + 5 =", total)

# You can also use the function directly in expressions
print("3 + 7 =", add(3, 7))
print("Sum of sums:", add(10, 20) + add(30, 40))
````

````python
# Without return, the function returns None
def greet_no_return(name):
    print(f"Hello, {name}!")

result = greet_no_return("Alice")
print("Return value:", result)  # None!
````

````python
# Practical example: area calculator
def rectangle_area(length, width):
    """Calculate the area of a rectangle."""
    return length * width

def circle_area(radius):
    """Calculate the area of a circle."""
    pi = 3.14159
    return pi * radius ** 2

# Using the functions
room = rectangle_area(5, 4)
print(f"Room area: {room} sq meters")

pizza = circle_area(15)
print(f"Pizza area: {pizza:.1f} sq cm")
````

## 📖 Default Parameters

You can give parameters **default values**. If the caller doesn't provide an argument, the default is used.

````python
# Default parameter values
def greet(name, greeting="Hello"):
    """Greet someone with a customisable greeting."""
    print(f"{greeting}, {name}!")

greet("Alice")              # Uses default greeting: "Hello"
greet("Bob", "Good morning")  # Overrides with "Good morning"
greet("Charlie", "Hey")
````

````python
# Practical example: tax calculator with default rate
def calculate_tax(amount, tax_rate=0.18):
    """Calculate tax. Default rate is 18% (GST)."""
    tax = amount * tax_rate
    return tax

print("Tax on ₹1000 at default rate:", calculate_tax(1000))       # 180.0
print("Tax on ₹1000 at 28%:", calculate_tax(1000, 0.28))          # 280.0
print("Tax on ₹1000 at 5%:", calculate_tax(1000, 0.05))           # 50.0
````

## 📖 Scope — Where Variables Live

**Scope** determines where a variable can be accessed:

| Scope | Where it's created | Where it can be used |
|-------|-------------------|---------------------|
| **Local** | Inside a function | Only inside that function |
| **Global** | Outside any function | Anywhere in the program |

> **Analogy:** A local variable is like a note on your desk — only you can see it. A global variable is like a notice on the office whiteboard — everyone can see it.

````python
# Local scope — the variable 'x' only exists inside the function
def my_function():
    x = 10  # Local variable
    print("Inside function: x =", x)

my_function()
# print(x)  # ❌ This would cause an error — x doesn't exist here!
````

````python
# Global vs Local
message = "I am global"  # Global variable

def show_message():
    message = "I am local"   # Local variable — DIFFERENT from the global one!
    print("Inside function:", message)

show_message()
print("Outside function:", message)  # Still the global value!
````

````python
warning_box("Avoid Global Variables",
    "While global variables work, they make code harder to understand and debug.<br>"
    "Best practice: Pass values as <b>parameters</b> and use <b>return</b> to get results back.<br>"
    "This keeps functions self-contained and predictable.")
````

## 📖 `*args` and `**kwargs` — Flexible Parameters

Sometimes you don't know in advance how many arguments your function will receive.

````python
# *args — accepts any number of positional arguments
def add_all(*numbers):
    """Add up any number of values."""
    total = 0
    for num in numbers:
        total += num
    return total

print(add_all(1, 2))           # 3
print(add_all(1, 2, 3, 4, 5))  # 15
print(add_all(10, 20, 30))     # 60
````

````python
# **kwargs — accepts any number of keyword arguments
def print_info(**details):
    """Print any key-value details."""
    for key, value in details.items():
        print(f"  {key}: {value}")

print("Student Profile:")
print_info(name="Alice", age=25, course="Data Science", city="London")
````

## 📖 Lambda Expressions — Mini Functions

A **lambda** is a small anonymous (unnamed) function written in one line.

```python
lambda parameters: expression
```

Lambdas are useful for simple operations you don't want to write a full function for.

````python
# Regular function
def double(x):
    return x * 2

# Same thing as a lambda
double_lambda = lambda x: x * 2

print("Regular:", double(5))
print("Lambda:", double_lambda(5))
````

````python
# Lambdas are often used with built-in functions
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

# Sort using a lambda as the key
sorted_nums = sorted(numbers)
print("Sorted:", sorted_nums)

# Sort strings by length
words = ["banana", "apple", "fig", "cherry"]
sorted_by_length = sorted(words, key=lambda w: len(w))
print("By length:", sorted_by_length)
````

    if len(password) < 8:
        return False
    has_digit = False
    has_upper = False
    for char in password:
        if char.isdigit():
            has_digit = True

## 🎯 Key Takeaways

1. **Functions** are reusable code blocks defined with `def function_name():`
2. **Parameters** receive input; **arguments** are the actual values passed in
3. **`return`** sends a value back to the caller; without it, the function returns `None`
4. **Default parameters** provide fallback values: `def f(x, y=10):`
5. **Local variables** only exist inside their function; **global variables** exist everywhere
6. **`*args`** accepts variable positional arguments; **`**kwargs`** accepts variable keyword arguments
7. **Lambda** expressions create small inline functions: `lambda x: x * 2`
